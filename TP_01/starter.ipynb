```python
#FactPix: Function to modify the intensenties of an image by multiplying with two factors.
def FactPix(Imagepath, factor1, factor2):
      image = cv2.imread(Imagepath)
      if len(image.shape)>2:
         image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
      #Normalize pixel values to range instead of 255 to a  range of [0,1]
      image = image/ 255.0
      I_fact = image * factor1 * factor2
      return I_fact

#utilisation
FactPix("/content/figs/flower.png",5,8)
```

```python
#Func-a :function that returns, logarithme, exponentielle, squared and the root of the entered image
def Func_a(Imagepath):
   image =plt.imread(Imagepath)
   img_exp = np.exp(image)
   img_log = np.log(image)
   squared_image = np.square(image)
   root_image = np.sqrt(image)
   return img_exp, img_log, squared_image, root_image

#utilisation de la fonction
img_exp, img_log, squared_image, root_image = Func_a("/content/figs/flower.png")

```

```python
#Func_m: Function to compute the mean and standard deviation of pixel intensities in an image
def Func_m(Imagepath):
    image = cv2.imread(Imagepath)
    mean = np.mean(image)
    std = np.std(image)

    return mean,std
print(Func_m("/content/figs/flower.png"))
```

```python
def Normalize(Imagepath,new_min=0,new_max=255):
    img = plt.imread(Imagepath)

    current_min = np.min(img)
    current_max = np.max(img)

    Norm_img = (image - current_min)*((new_max - new_min) / (current_max - current_min)) + new_min

    return Norm_img
Inorm = Normalize("/content/figs/flower.png",new_min=0, new_max=1.5)
img = cv2.imread("/content/figs/flower.png")
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(Inorm)
plt.title("Normalized Image")
plt.show()
```

```python
#Inverse: Function to compute the inverse (negative) of an image.
def Inverse(Imagepath):
    image=cv2.imread(Imagepath)
    Inversed_img = np.max(image) - image
    return Inversed_img

#utilisation de la focntion
Inv = Inverse("/content/figs/flower.png")

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(Inv)
plt.title("Inversed Image")
plt.show()
```

```python
#CalcHist: Function to calculate the histogram of an image.
from PIL import Image
def CalcHist(Imagepath):
   img = cv2.imread(Imagepath)

   #reload the image in grayscale
   img = Image.open(Imagepath).convert("L")

   #convert the image to float dtype with a value range from 0 to 1
   img = np.array(img) / 255.0

   H, bins = np.histogram(img, bins=256,range=(0,1))

   return H, bins

histogram, bin_edges = CalcHist("/content/figs/flower.png")

```

```python
# Plot the histogram
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(bin_edges[:-1], histogram)
plt.title('Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

# Plot the image
plt.subplot(1, 2, 2)
plt.imshow(img, cmap='gray')
plt.title('Image')

plt.show()
```

```python
# Threshold: Function to perform thresholding on an image.
def Threshold(Imagepath, threshold):
  image = plt.imread(Imagepath)

  T = np.where(image > threshold, 255, 0)
  return T
#utilisation de la fonction

threshold_value = 50
thresholded_image = Threshold("/content/figs/flower.png",threshold_value)

```

```python
import matplotlib.pyplot as plt

# Define pixel values
BLACK = 0
GRAY = 128
WHITE = 255
```

```python
# Create a checkered image (modify size as needed)
check_size = 8  # Size of each checker square
```

```python
# Create empty image
image_size = 32  # Modify image size as needed
image = np.full((image_size, image_size), WHITE, dtype=np.uint8)

# Fill the image with alternating black and white squares
for i in range(0, image_size, check_size):
    for j in range(0, image_size, check_size):
        # Check if the square is even or odd numbered (to alternate colors)
        if (i // check_size) % 2 == (j // check_size) % 2:
            image[i:i+check_size, j:j+check_size] = BLACK
        else:
            image[i:i+check_size, j:j+check_size] = GRAY

# Visualize the image
plt.imshow(image, cmap="gray")
plt.title("Checkered Image")
plt.show()
```
